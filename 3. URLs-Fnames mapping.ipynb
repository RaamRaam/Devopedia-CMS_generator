## Removing Duplicate Rows

```python
x=set()
```

```python
for i in df_references.iterrows():
    x.add(i[0])
```

```python
len(x)
```

```python
dicto={'url':[],'fname':[],'text':[],'citeas':[],'file_type':[],'file_size':[]}

for i in x:
    try:
        dicto['fname'].append(df_refs.fname[i])
        dicto['url'].append(i)
        dicto['text'].append(df_references.text[i])
        dicto['citeas'].append(df_references.citeas[i])
        dicto['file_type'].append(df_refs.type[i])
        dicto['file_size'].append(df_refs['size'][i])
    except:
        continue
    
```

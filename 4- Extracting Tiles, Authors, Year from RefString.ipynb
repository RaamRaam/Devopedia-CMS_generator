```python
import re
import pandas as pd
```

```python
df=pd.read_csv('Dataset_deduplicated.csv')
```

```python
df
```

### Removing invalid texts (starting with 'url')

```python
error_url_indices=[]

for i in df.iterrows():
    x=i[1].text
    url_row=re.match(r'url',x)
    if url_row:
        error_url_indices.append(i[0])
```

```python
for i in error_url_indices:
    df.drop(i,inplace=True)
```

```python
df=df.reset_index()
df.drop(['index'],axis=1,inplace=True)
```

```python
df.shape
```

## Creating Dates Column

```python
Dates=[]
for i in df.text:
    date_reg=re.search('\d\d\d\d',i)            #Extracting the first 4-digit number (year) from the Reference String
    date=int(date_reg.group())                  #String to Int
    Dates.append(date)
```

```python
len(Dates)
```

```python
df['Date']=Dates
```

## Creating Titles Column

```python
Titles=[]
errors=[]
for index,i in enumerate(df.text):
    try:
        print(index)
        title=re.search('"([^"]*)"',i)                #Text within double quotes in the RefString is the Title
        title=title.group()
        title=title.replace('"','')                   #Removing double quotes from the text
        print(title)
        Titles.append(title)
        print()
    except:
        Titles.append(None)
        errors.append(index)
```

```python
print(len(errors))
print(len(Titles))
```

### Files causing errors

```python
for i in errors:
    print(df.text[i])
    print()
```

```python
df['Title']=Titles
```

```python
df=df.dropna()       #Dropping NULL titles (previously appended due to tracebacks)
df=df.reset_index()
df.drop(['index'],axis=1,inplace=True)
```

```python
df.shape
```

## Creating Authors Column

```python
authors=[]

for i in df.text:
    author=re.search(r'^([^.]+)',i)
    author=author.group()
    authors.append(author)
```

```python
len(authors)
```

```python
df['Authors']=authors
```

```python
df.to_csv('Dataset_labelled.csv',index=False)
```

# Creating a DataFrame of URLs and their corresponding Reference Strings

```python
import os
import pandas as pd
from tqdm import tqdm
```

```python
ArticleFile=r'C:\Users\Saranga\Desktop\Devopedia\Work\Devopedia\devopediaArticles.v2525.json'
```

```python
ARTICLES = pd.read_json(ArticleFile, orient='index')
```

```python
ARTICLES
```

```python
ARTICLES=ARTICLES.drop(['alias','version'],axis=1)
```

```python
#copied from provided Github repo's Devopedia-Article-Reader.ipynb file

secs = ARTICLES['secs'].apply(pd.Series)
ARTICLES = pd.concat([ARTICLES, secs], axis=1)
articles = ARTICLES[~ARTICLES.References.isna()]
```

```python
articles
```

```python
dicto={'url':[],'text':[],'citeas':[]}
```

```python
for item in articles.References:
    for i in tqdm(item):
        try:
            dicto['url'].append(i['url'])
            dicto['citeas'].append(i['citeas'])
            dicto['text'].append(i['text'])
        except:
            continue
```

```python
#Verifying wheteher the Columns have the same number of rows

print(len(dicto['url']))
print(len(dicto['text']))
print(len(dicto['citeas']))
```

```python
df=pd.DataFrame.from_dict(dicto)
```

```python
df.to_csv('URL_RefStrings.csv',index=False)
```

# From Labelled dataset, consider only html files

```python
import pandas as pd
import os
from tqdm import tqdm
```

```python
df=pd.read_csv('dataset_labelled.csv')
```

```python
df
```

```python
df_html=df
```

```python
for i in df.file_type.unique():
    print(i)
```

```python
exclude_file_types=['text/plain; charset=UTF-8','application/pdf','application/x-zip-compressed','application/octet-stream',
                   'application/pdf;charset=ISO-8859-1','application/pdf; charset=binary','application/pdf; charset=UTF-8',
                   'application/vnd.ms-powerpoint','text/plain; charset=utf-8','text/html; charset=EUC-JP','text/plain; charset=utf-8',
                   'text/plain','image/jpeg','application/pdf; charset=','application/pdf;charset=utf-8','application/pdf;charset=UTF-8',
                   'image/svg+xml','application/pdf;charset=','image/png','image/svg+xml; qs=0.85','application/postscript']
```

```python
for i in df.file_type.unique():
    if i not in exclude_file_types:
        print(i)
```

```python
for i in df.iterrows():
    if i[1]['file_type'] in exclude_file_types:
        df_html=df_html.drop(i[0])

df_html.reset_index(inplace=True)
```

```python
df_html.to_csv('dataset_html.csv',index=False)
```
